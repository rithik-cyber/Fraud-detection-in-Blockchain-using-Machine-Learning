```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('transaction_dataset.csv')
df
```

```python
df.describe()
```

```python
missing_values = df.isnull().sum()
data_types = df.dtypes
df_cleaned = df.drop(columns=['Unnamed: 0', 'Index'])
missing_values, data_types, df_cleaned.head()
```

**Pre-processing**


```python
# Impute missing values for ERC20 features with appropriate techniques (e.g., mean, median, or 0, based on the context).
# Features with large ranges (like Avg min between sent tnx, Max_Value_Received, etc.) need to be normalized or standardized to ensure consistency.
df_cleaned.fillna(0, inplace=True)
from sklearn.preprocessing import MinMaxScaler
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
scaler = MinMaxScaler()
df_cleaned[numeric_cols] = scaler.fit_transform(df_cleaned[numeric_cols])
df_cleaned.head()
```

***Exploratory Data Analysis***

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(x='FLAG', y='Avg min between sent tnx', data=df_cleaned, ax=ax[0])
ax[0].set_title('Average Time Between Sent Transactions')
ax[0].set_xlabel('Fraud Flag (0 = Non-Fraud, 1 = Fraud)')
ax[0].set_ylabel('Avg Min Between Sent Transactions')

sns.boxplot(x='FLAG', y='Avg min between received tnx', data=df_cleaned, ax=ax[1])
ax[1].set_title('Average Time Between Received Transactions')
ax[1].set_xlabel('Fraud Flag (0 = Non-Fraud, 1 = Fraud)')
ax[1].set_ylabel('Avg Min Between Received Transactions')

plt.tight_layout()

# Save the plot in the current working directory
plt.savefig('fraud_boxplots.png', dpi=300)

# Display the plot
plt.show()

```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

# Initialize the models
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
logistic_model = LogisticRegression(random_state=42)
xgb_model_optimized = XGBClassifier(n_estimators=50, use_label_encoder=False, eval_metric='logloss', scale_pos_weight=10, random_state=42)

# Define cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)  # 5-fold stratified cross-validation

# Ensure X_cleaned_aligned and y_train are your feature and target variables respectively
# X_cleaned_aligned: Your features (2D array, shape: [n_samples, n_features])
# y_train: Your target variable (1D array, shape: [n_samples])

# Perform cross-validation for each model
rf_scores = cross_val_score(rf_model, X_cleaned_aligned, y_train, cv=cv, scoring='roc_auc')
xgb_scores = cross_val_score(xgb_model_optimized, X_cleaned_aligned, y_train, cv=cv, scoring='roc_auc')
logistic_scores = cross_val_score(logistic_model, X_cleaned_aligned, y_train, cv=cv, scoring='roc_auc')

# Compute the mean ROC AUC for each model
rf_mean_auc = rf_scores.mean()
xgb_mean_auc = xgb_scores.mean()
logistic_mean_auc = logistic_scores.mean()

# Print the results
print(f"Random Forest Mean ROC AUC: {rf_mean_auc}")
print(f"XGBoost Mean ROC AUC: {xgb_mean_auc}")
print(f"Logistic Regression Mean ROC AUC: {logistic_mean_auc}")
```
